```python
from collections import deque # python의 고유 built-in library

# 1. 큐 선언과 원소 추가를 해보자

queue=deque() # 큐 선언

queue.append(5) # 차례대로 원소 추가
queue.append(3)
queue.append(7)
queue.append(2)
queue.append(8)

# 2. 큐에서 원소 찾기

step=1
for i in range(len(queue)):
    p=queue.popleft()
    print(f'단계 {step}: 큐에서 원소 {p} 제거')
    step+=1

# 3. 큐의 가장 앞에 있는 원소 확인하기

queue2=deque()

queue2.append(4)
queue2.append(6)
queue2.append(9)
queue2.append(1)
queue2.append(7)

if len(queue2)>0:
    front=queue2[0]
    print('\nfront item:',front)
```

```python
# GRAPH 연습 1
graph1 = {
    5:[3,7],
    7:[2,3,5],
    3:[5,2,1],
    1:[2,3],
    2:[1,7]
}

# GRAPH 연습 2
graph2 = {
    4:[9],
    9:[4,3,7],
    3:[7],
    7:[],
    1:[7]
}

# GRAPH 연습 3
graph3 = {
    'A':['B','D'],
    'B':['A','C'],
    'C':['B','F'],
    'D':['A','E'],
    'E':['D','F'],
    'F':['E','C']
}

# Node의 개수 구하기
numnode=len(graph3)
print(numnode)

# Edge의 개수 구하기 (total connection)
totaledge=0
for node in graph3.values():
    totaledge+=len(node)

numedge=totaledge//2 # slash 2개는 '몫' 구하는거!! 참고쓰
print('실제 연결수는:', numedge)

# List Comprehension: Type 1
edgenum_lst=[len(n) for n in graph3.values()]
print('각 values들의 길이들',edgenum_lst,'\n') # [2, 2, 2, 2, 2, 2] --> values의 길이들

totaledge2=sum(edgenum_lst)//2
print('연결된 수',edgenum_lst,'\n')

# List Comprehension: Type 2
connection3=sum(len(n) for n in graph3.values())//2 ; print('연결된 수',connection3)
```

```python
# List conprehension --> 코드를 한줄로 줄이기!

# 원래 코드
print('List 1: Original Code')
lst1=[]

for i in range(1,11):
    if i%2==0:
        lst1.append(i)

print(lst1,'\n')

# Comprehensed list
print('List 2: Comprehensed List')
lst2=[i for i in range(1,11) if i%2==0]
print(lst2)
```
